## Exploratory Data Analysis

En este archivo nos centraremos en analizar nuestros datos, revisar los datasets proporcionados y tratar de decifrar que contienen y que acciones tomar para poder mejorar nuestros datos.

```python
#Empezamos importando las librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Nuestro dataset se divide en dos partes, la primera que son los hechos y la segunda que es una hoja de victimas. Desde el repositorio principal importamos nuestros datos en formato .csv para poder leerlo con pandas desde un principio, entonces tenemos dos archivos .csv con todos nuestros datos.

```python
#Importamos los datos con formato ISO-8859-1 para su correcta lectura en pandas.
hechos_df = pd.read_csv('../Data/HechosEDA.csv', encoding= 'ISO-8859-1')
victimas_df = pd.read_csv('../Data/VictimasEDA.csv', encoding= 'ISO-8859-1')
```

```python
#Visualisamos los dataframes superficialmente
hechos_df.head()
```

```python
victimas_df.head()
```

```python
#Vemos su información
hechos_df.info()
```

Se puede observar que tenemos datos de tipo str, enteros y flotantes. Tenemos datos de fechas, direcciones de donde fue el accidente, la cantidad de victimas, los involucrados y el tipo de accidente.


### Gráficas

Para observaciones, vamos a graficar las victimas y los acusados para poder encontrar algun insight.

```python
sns.histplot(data=hechos_df['VICTIMA'])
plt.xticks(rotation= 90)
plt.ylabel(ylabel= 'Cantidad De Victimas')
plt.show()
```

Podemos observar que el top 3 de victimas respectivamente son:

- Moto
- Peaton
- Auto

```python
sns.histplot(data=hechos_df['ACUSADO'])
plt.xticks(rotation= 90)
plt.ylabel(ylabel= 'Cantidad De Acusados')
plt.show()
```

Podemos observar que el top 3 de acusados son:
- Auto
- Pasajeros
- Cargas

```python
#Vamos a hacer el display de las comunas en donde hasy más accidentes
sns.barplot(data= hechos_df['COMUNA'].value_counts())
plt.xlabel(xlabel= 'Comuna')
plt.ylabel(ylabel= 'Cantidad De Accidentes')
plt.xticks(rotation= 90)
plt.show()
```

Aquí podemos observar el top de comunas que están presentes en la mayoría de accidentes, donde destacan:
- La coumna 1
- La comuna 4
- La comuna 9

```python
#Ahora vamos a analizar el año que más accidentes tuvo
sns.histplot(data= hechos_df['AAAA'])
plt.xlabel(xlabel= 'Año')
plt.ylabel(ylabel= 'Cantidad de Accidentes')
```

Al parecer el top 3 esta en los primeros 3 años, pero luego despues del 2018 los incidentes bajan y luego suben en 2021. Entonces el top 3 es:

- 2016
- 2018
- 2017

```python
#Ahora vamos a obbservar si existen datos nulos
nulos = hechos_df[hechos_df.isnull().any(axis= 1)]
nulos
```

Al parecer contamos con 696 filas que contienen datos nulos, pero al parecer solo en la columna de altura. No parece que esta información sea tan necesaria para nuestro analisis, por lo cual no se eliminaran esas filas y solamente se les colocará una altura "x" en la parte del ETL.

Ahora nos centramos en el dataset de victimas, este dataset se sentra en mostrar los datos de las victimas de los accidentes.

Aquí podemos observar que no es un gran dataset, pero al parecer tenemos un valor en Fecha de fallecimiento que es SD, esto significa Sin Dato. Observemos cuantos SD tenemos.

```python
victimas_df.head()
```

```python
sd_victimas = victimas_df.applymap(lambda x: 'SD' in str(x))
sd_rows = victimas_df[sd_victimas.any(axis= 1)]
sd_rows
```

Podemos ver que contamos con 109 filas que no tienen dato, es el 15% de nuestra información, eliminarla sería una gran perdida, por lo tanto se obtará por asignar datos a estas columnas para no perder la demás información.

Otra cosa importante a notar es que tenemos fechas como strings completos, por lo tanto en la EDA optaremos por separar esas fehcas en año, mes y día para tener dicha info más accesible.

```python
#Ahora para observar los datos, vamos a analizar en que año existieron más victimas
sns.histplot(data= victimas_df['AAAA'])
plt.xlabel(xlabel= 'Año')
plt.ylabel(ylabel= 'Cantidad de Victimas')
plt.show()
```

Como era de esperar el mayor número de vicitmas también se encuentra en los años con mayor número de accidentes.

```python
#Un dato interesante es ver el sexo de la victima, primero vamos a asegurarnos que todos los valores solo sean femenino y masculino
victimas_df['SEXO'].unique()

#Tenemos sexo masculino, femenino y SD (valores faltantes). Ahora vamos a graficarlos:
sns.histplot(data= victimas_df['SEXO'])
plt.xlabel(xlabel= 'Sexo')
plt.ylabel(ylabel= 'Cantidad')
```

Facilmente podemos observar que el top es el **sexo masculino**.

```python
sns.violinplot(data= victimas_df, x= victimas_df['VICTIMA'], y= victimas_df['ROL'])

plt.xticks(rotation= 90)
plt.show()
```

Como podemos observar de esta gráfica, para moto y auto el mayor porcentaje de victimas están cuando la victima es el conductor. Para los vehiculos de carga se mantiene casi igual si es acompañante o conductor y para movil es mayor cuando es el acompañante.
